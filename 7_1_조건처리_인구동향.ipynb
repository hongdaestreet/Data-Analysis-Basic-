```python
# 연도에 해당하는 데이터프레임 만드는 함수
def makeDf(year):   # 파라미터: year 인수 전달
    col = [str(year) if i == 0 else str(year) + '.'+ str(i) for i in range(4)]
    col.insert(0,'행정구역별(1)')   # 첫번째 열에 행정구역 추가

    temp = data.loc[1:, col]    # 0번째 타이틀은 필요없어서 제거, 1~끝
    temp.columns = ['지역', '출생아수', '사망자수', '혼인건수', '이혼건수']
    temp['연도'] = year # 연도열 추가해서 매개변수 year로 표시

    return temp

makeDf(2018)
```

• 행인덱스 재설정</br>
✓ df.reset_index(drop=False, inplace=False)</br>
➢ drop : index를 삭제할지 결정, True면 삭제</br>
➢ inplace : 원본을 변경할지 결정, True면 변경

```python
# 연도별 데이터 합치기, 행인덱스 재설정
df = makeDf(2018)

for y in range(2019,2023):
    df = pd.concat([df, makeDf(y)])

df.reset_index(drop=True, inplace=True)
df
```

```python
# 열 타입 변경
df = df.astype({'출생아수': int,'사망자수':int, '혼인건수':int, '이혼건수': int})   # 딕셔너리 형태
df.info()
```

```python
# 연도별 전국 자료  -> 이거 이상함
df1 = df[df['지역'] == "전국"]
df1
```

```python
# 연도별 전국 자료: 연도 숫자 타입을 문자로 변경 후 연도열을 행인덱스로 변경
# df1 = df.query('지역 == "전국"')
df1['연도'] = df1['연도'].astype('str')
df1 = df1.set_index('연도')
df1
```

```python
# 열 일부 내용만 추출
df1 = df1.loc[:, ['출생아수','사망자수','혼인건수']]
df1
```

```python
# 자연증가수 계산
df1['자연증가수'] = df1['출생아수'] - df1['사망자수']
df1
```

```python
fig, ax1 = plt.subplots(figsize =(8,4))
ax1.plot(df1.index, df1['출생아수'], 'b--o',label='출생아수')
ax1.plot(df1.index, df1['사망자수'], 'g--o',label='사망자수')
plt.legend()
plt.title('연도별 인구동향')

ax2= ax1.twinx()
ax2.plot(df1.index,df1['자연증가수'], 'r-x')
ax2.axhline(y=0, color='r',ls='--',lw=0.5)

plt.show()
```

지역별 자료

```python
# 지역별 자료: 전국이 아닌 자료 추출
df2 = df.query('지역 != "전국"')
df2
```

```python
# 혼인건수와 출생아수 관계
import seaborn as sns
df2['연도'] = df2['연도'].astype(str)   # 범주를 str로 수정 시 색상이 바뀜
sns.scatterplot(x='혼인건수', y='출생아수', hue='연도', data = df2)
plt.title('혼인건수와 출생아수 관계')
plt.show()
```

```python
# 평균과 표준편차 표시
sns.barplot(x='지역', y='출생아수', data=df2)
plt.title('지역별 출생아수 평균')
plt.show()
```

```python
# 다른 방식
# 평균과 표준편차 표시
sns.barplot(x='지역', y='출생아수', data=df2.sort_values('출생아수', ascending=False))
plt.title('지역별 출생아수 평균')
plt.show()
```

• 그룹화</br>

✓ 데이터를 그룹별로 분류</br>

✓ df.groupby(인수)</br>
➢ 열 또는 열의 리스트</br>
➢ 행 인덱스</br>

✓ 그룹 연산 메소드</br>
➢ size.count : 그룹 데이터의 개수</br>
➢ mean, median, min, max : 평균, 중앙값, 최소, 최대</br>
➢ sum, prod, std, var, quantile : 합, 곱, 표준편차, 분산, 사분위수</br>
➢ first, last : 첫번째, 마지막

```python
dfg = df2.groupby('지역').mean()
dfg
```

```python
dfg[['출생아수', '사망자수']].sort_values('출생아수', ascending=False)
```

```python
dfg[['출생아수', '사망자수']].sort_values('출생아수', ascending=False).plot.bar()
dfg[['혼인건수', '이혼건수']].sort_values('혼인건수', ascending=False).plot.bar()
```
