작성완료

• HTTP 데이터 전송 방법</br>
✓ GET 방식</br>
➢ 브라우저에서 주소창에 URL 입력하면 전송</br>
➢ URL 끝에 요청 데이터를 붙여서 전송, 보통 데이터가 노출됨</br>
➢ 요청 데이터의 크기가 제한적임</br>
✓ POST 방식</br>
➢ 요청 데이터를 HTTP 메시지 본문에 넣어서 전송</br>
➢ URL에 데이터가 노출되지 않으며, 보안성이 높음</br>
➢ 데이터 크기 제한이 없음</br>

• 비교 : HTTP POST는 데이터를 서버에 전송하는 방법 중 하나, REST API의 POST 메소드는 서버에 데이터를 생성하는 메소드</br>

• JSON 모듈</br>
✓ JavaScript Object Notation</br>
➢ 데이터 교환 포맷으로 키-값의 쌍으로 이루어진 컬렉션</br>
➢ 데이터 포맷이 단순하고 유연함</br>
➢ 웹 상에서 데이터를 교환하는 데 많이 사용 중</br>
✓ Python JSON 표준 라이브러리</br>
➢ import json</br>

• JSON 인코딩</br>
✓ python 타입의 데이터를 JSON 문자열로 변경</br>
✓ 딕셔너리, 리스트, 튜플을 JSON 문자열로 변경</br>
✓ 반환값은 JSON 표현을 갖는 문자열(str 타입)</br>
✓ “indent” 옵션</br>
➢ JSON 문자열을 읽기 편하게 indentation이 적용된 문자열 반환</br>
➢ json.dumps(파이썬자료, indent=4)</br>
✓ 한글 처리</br>
➢ ensure_ascii=False</br>
➢ json.dumps(파이썬자료, ensure_ascii=False)</br>

• JSON 디코딩</br>
➢ json.loads()

```python
# POST 방식의 json 포맷으로 전송
import requests
import json
```

```python
indata = input('질문을 입력하세요')
```

```python
# HTTP 요청 헤더 및 바디 설정 : dics 형태로 표시
url = 'http://aiopen.etri.re.kr:8000/WikiQA'    # HTTP 요청을 보낼 대상의 URL 저장

# HTTP 요청에 대한 부가 정보를 전달
headers = {
    "Content-Type": "application/json; charset=UTF-8",
    "Authorization": 'a7a2d8db-8a85-4435-b0b6-da15f5e4a2c4'
    }

# HTTP 요청 본문
body = {
    'argument': {
        "type": 'ENGINE_TYPE',  #  질문 유형
        "question": indata   # input 사용시 indata, '질문 가능'
    }
}
```

```python
# HTTP 요청 보내기
resp = requests.post(url, headers=headers, data=json.dumps(body))   # json 형태의 모듈
```

```python
# @title 기본 제목 텍스트
# 응답 결과
data = resp.json()  # 딕셔너리 형태로 변환
data
```
