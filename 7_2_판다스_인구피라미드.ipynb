## 합계 출산율 (project2)

```python
df = pd.read_excel('/content/gdrive/MyDrive/PNU/3-2/데이터분석입문(겨울)/합계출산율.xlsx')
df
```

```python
df.index
```

```python
df.index.values
```

```python
df = df.set_index('Unnamed: 0')
df
```

```python
df.index
```

```python
df.index.values
```

```python
df.rename(index={'출생아\xa0수': '출생아 수', '합계\xa0출산율': '합계 출산율'}, inplace=True)
df.index.values
```

```python
df.loc['출생아 수'] # df.iloc[0] 동일함
```

```python
# 구조화 데이터
dft = df.T
dft
```

```python
plt.plot(dft.index, dft['출생아 수'])
plt.plot(dft.index, dft['합계 출산율'])
plt.show()
```

```python
# x축 공유
fig, ax1 = plt.subplots(figsize = (10,7))
ax1.plot(dft.index, dft['출생아 수'], color='#ff812d')
ax2 = ax1.twinx()   # X축을 공유하는 쌍둥이 axis 생성, Y 범주 변경
ax2.plot(dft.index, dft['합계 출산율'], color = '#ffd100')
plt.show()
```

```python
# x축 공유
fig, ax1 = plt.subplots(figsize = (10,7))
ax1.set_ylabel('출생아 수 (천명)')
ax1.bar(dft.index, dft['출생아 수'], color='#ff812d')
ax2 = ax1.twinx()   # X축을 공유하는 쌍둥이 axis 생성, Y 범주 변경
ax2.set_ylabel('합계 출산율 (가임여성 1명당/명)')
ax2.plot(dft.index, dft['합계 출산율'], color = '#ffd100')
plt.show()
```

```python
# 혼합그래프
fig, ax1 = plt.subplots(figsize=(13,5))
fig.suptitle('출생아 수 및 합계출산율') # 전체 그림에 제목 추가
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(200, 600)      # y축 범위
ax1.set_yticks([200,300,400,500])   # y축 눈금 설정
ax1.bar(dft.index, dft['출생아 수'], color='#ff812d')
for idx, val in enumerate(dft['출생아 수']):
    ax1.text(idx, val + 10, val, ha = 'center')

ax2 = ax1.twinx()   # 첫 번째 축과 동일한 x축 공유
ax2.set_ylabel('합계 출산율 (가임여성 1명당/명)')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])
ax2.plot(dft.index, dft['합계 출산율'], color='#ffd100', marker = 'o', mec = 'w', mew = 3, ms = 10, lw = 3)
for idx, val in enumerate(dft['합계 출산율']):
    ax2.text(idx, val + 0.05, val, ha = 'center')
plt.show()
```
